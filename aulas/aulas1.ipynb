```python
class Funcionario:
    def __init__(self,nome,salario):
        self._nome = nome
        self._salario = salario

    @property
    def get_salario(self):
        return self._salario
    
    @property
    def get_nome(self):
        return self._nome
    
    def __str__(self):
        return f'Nome: {self._nome} | Salário: {self._salario}'   
    
class FuncionarioCLT(Funcionario):
    def __init__(self,nome,salario):
        super().__init__(nome,salario)
        self.imposto = 22
        
    def calcular_imposto(self):
        desconto = self._salario * (self.imposto / 100)
        return desconto
    
    def get_infos(self):
        return f'Nome: {self._nome} | Salário: R${self._salario:.2f}| Imposto: R${self.calcular_imposto():.2f}'   
    
class FuncionarioPJ(Funcionario):
    def __init__(self,nome,cnpj,salario,imposto):
        super().__init__(nome,salario)
        self.__cnpj = cnpj
        self.__imposto = imposto


    def calcular_imposto(self):
        return self._salario * (self.__imposto / 100)

    def get_infos(self):
        return f'Nome: {self._nome} | CNPJ: {self.__cnpj} | Salário: R${self._salario:.2f}| Imposto: R${self.calcular_imposto():.2f}'   
```

```python
clt = FuncionarioCLT('João', 3000)
pj = FuncionarioPJ('Maria', '12.345.678/0001-90', 5000, 15)
print(clt.get_infos())
print(pj.get_infos())
```
