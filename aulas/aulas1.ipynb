```python
class Estoque:
    def __init__(self):
        self.produtos = []

    def adicionar_estoque(self,produto):
        if self.buscar_estoque(produto.nome):
            return self.repor(produto.nome,produto.estoque)
        else:
            self.produtos.append(produto)
            return f'O produto {produto.nome} foi adicionado ao estoque. Capacidade atual: {produto.estoque} unidades.'

    def buscar_estoque(self, nome ):
        for produto in self.produtos:
            if produto.nome == nome:
                return produto
        return None

    def remover_estoque(self):
        for produto in self.produtos:
            if produto.estoque == 0:
                self.produtos.remove(produto)
                return f'O produto {produto.nome} foi removido do estoque por falta de unidades.'

    def verificar_estoque(self):
        if self.produtos:
            return 'Produtos em estoque: ' + ', '.join([produto.nome for produto in self.produtos])
        else:
            return 'Estoque indisponivel. Nenhum produto encontrado.'       

    def listar_estoque(self):
        for produto in self.produtos:
            print(produto)

    def total_estoque(self):
        total = sum(produto.preco * produto.estoque for produto in self.produtos)
        return f'Total do valor em estoque: R${total:.2f}'
    
    def repor(self, nome, quantidade):             
        produto = self.buscar_estoque(nome)        
        if produto and quantidade > 0:
            return produto.repor(quantidade)       
        return 'Produto não encontrado ou quantidade inválida.'

class Produto:

    def __init__(self,nome,preco,estoque):
        super().__init__()
        self.nome = nome
        self.preco = preco
        self.estoque = estoque

    def vender(self,quantidade):
        if quantidade > 0:
            if quantidade <= self.estoque:
                self.estoque -= quantidade
                return f'Foram vendidas {quantidade} unidades de {self.nome} no valor de R${self.preco * quantidade:.2f}.'
            else:
                return f'Não há estoque suficiente para vender {quantidade} unidades de {self.nome}.'
        else:
            return 'A quantidade de venda deve ser maior que zero.' 
        
    def repor(self,quantidade):
        if quantidade > 0:
            self.estoque += quantidade
            return f'{quantidade} unidades de {self.nome} foram repostas ao estoque. Capacidade atual: {self.estoque} unidades.'           
        else:
            return 'A quantidade de reposição deve ser maior que zero.'    

        
    def __str__(self):
        return f'Produto: {self.nome} | Preço unitario: R${self.preco:.2f} | Estoque: {self.estoque} unidades'
        

class ProdutoPerecivel(Produto):
    def __init__(self,nome,preco,estoque,data_validade):
        super().__init__(nome,preco,estoque)
        self.data_validade = data_validade


    def verificar_validade(self):
        from datetime import datetime
        dt_validade = datetime.strptime(self.data_validade, '%Y-%m-%d').date()
        data_atual = datetime.now().date()
        dias_vencido = (data_atual - dt_validade).days
        if dt_validade < data_atual:
            return f'O produto {self.nome} está vencido a {dias_vencido} dias. Data de validade: {dt_validade}'
        else:
            return f'O produto {self.nome} está dentro do prazo de validade. Data de validade:{dt_validade}'  

    def __str__(self):
        return f'Produto: {self.nome} | Preço unitario: R${self.preco:.2f} | Estoque: {self.estoque} unidades | Data de validade: {self.data_validade}'          


```

```python
estoque = Estoque()

teclado = Produto("Teclado", 150.00, 10)
print(estoque.adicionar_estoque(teclado))   # adiciona

teclado2 = Produto("Teclado", 150.00, 5)
print(estoque.adicionar_estoque(teclado2))  # deve repor +5

print(estoque.total_estoque())      
```

```python
class Funcionario:
    def __init__(self,nome,salario=0):
        self.nome = nome
        self._salario = salario

    @property
    def salario(self):
        return f'R${self._salario:.2f}'
    
    @salario.setter
    def dar_aumento(self,percentual):
        aumento = 1 + percentual / 100
        self._salario = self._salario * aumento

    
```

```python
f1 = Funcionario('Wellington',2500)
print(f1.salario)
f1.dar_aumento = 10
print(f1.salario)
```
